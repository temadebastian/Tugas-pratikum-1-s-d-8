```python
#Pseudocode
FUNGSI shortest_path(graf, start):

  UNTUK setiap node di graf:
    jarak[node] = TAK_TERHINGGA
    sebelumnya[node] = KOSONG

  jarak[start] = 0
  antrian = semua node

  SELAMA antrian tidak kosong:

    u = node di antrian dengan jarak TERKECIL
    hapus u dari antrian

    UNTUK setiap tetangga v dari u:
      jarak_baru = jarak[u] + bobot(u, v)

      JIKA jarak_baru < jarak[v]:
        jarak[v] = jarak_baru
        sebelumnya[v] = u

  KEMBALIKAN jarak[], sebelumnya[]
```

```python
import heapq

def shortest_path(graf, start):

    jarak = {node: float('inf') for node in graf}
    jarak[start] = 0
    sebelumnya = {node: None for node in graf}
    antrian = [(0, start)]

    while antrian:
        jarak_sekarang, u = heapq.heappop(antrian)

        for tetangga, bobot in graf[u]:
            jarak_baru = jarak[u] + bobot

            if jarak_baru < jarak[tetangga]:
                jarak[tetangga] = jarak_baru
                sebelumnya[tetangga] = u
                heapq.heappush(antrian, (jarak_baru, tetangga))

    return jarak, sebelumnya

graf = {
    'A': [('B', 4), ('C', 2)],
    'B': [('A', 4), ('D', 5)],
    'C': [('A', 2), ('B', 1), ('D', 8)],
    'D': [('B', 5), ('C', 8)],
}

jarak, sebelumnya = shortest_path(graf, 'A')

for node, d in sorted(jarak.items()):
    print(f"A -> {node} : {d}")
```

Kompleksitas waktu algoritma ini adalah O((V + E) log V), di mana V adalah jumlah titik dan E adalah jumlah sambungan antar titik.
Angka log V muncul karena kita pakai heap (heapq) sebagai antrian, yang otomatis ngurutin data dari yang terkecil. Setiap titik dan sambungan diproses cuma sekali, tapi tiap prosesnya butuh waktu log V karena heap perlu nyusun ulang urutannya.
Kalau dibandingkan, pakai list biasa kompleksitasnya jadi O(V²) karena harus ngecek semua titik satu-satu buat nyari yang terkecil. Makanya pakai heap jauh lebih efisien, apalagi kalau datanya banyak.
